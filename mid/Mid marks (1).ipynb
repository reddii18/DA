```python
df.rename(columns={'M-II':'M2'}, inplace=True)
df
```

```python
columns_to_convert = ['DV', 'M2', 'PP', 'BEEE', 'FL', 'FIMS']
for col in columns_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df['Total'] = df['DV'] + df['M2'] + df['PP'] + df['BEEE'] + df['FL'] + df['FIMS']
df
```

```python
df['PERCENTAGE'] = (df['Total'] /120)*100
print(df)
```

#Calculating percentage based on 'Total'column value

# Renaming column 'M-II' to 'M2' in dataframe

```python
def assign_grade(percentage):
    if percentage >= 90:
        return 'A'
    elif percentage >= 80:
        return 'B'
    elif percentage >= 70:
        return 'C'
    elif percentage >= 60:
        return 'D'
    else:
        return 'F'
df['GRADE'] = df['PERCENTAGE'].apply(assign_grade)
df

```
